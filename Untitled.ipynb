```python
import pandas as pd
import spacy
from spacy.lang.en import English 
```

```python
with open ('17-0.txt', "r") as f:
  text = f.read()
```

```python
nlp = English()
```

```python
doc = nlp(text[5000:7000])
```

```python
print (doc)
```

```python
for token in doc:
  print (token)
```

```python
from spacy import displacy
```

```python
displacy.render(doc, style = "ent")
```
